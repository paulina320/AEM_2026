# AEM 2026 – Notebook 3: Statistical Analysis

This notebook demonstrates the statistical-analysis tools in `aem`, using
the bundled sample datasets as running examples.

Topics covered:
1. Independent-samples *t*-test  
2. One-way ANOVA  
3. Effect sizes (Cohen's *d*, η²)  
4. Pearson correlation  
5. χ² test of independence  
6. Prospective power analysis  
7. Interaction plots

```python
import matplotlib.pyplot as plt
import aem

rt  = aem.load_sample('reaction_times')
mem = aem.load_sample('memory_recall')
caf = aem.load_sample('caffeine_alertness')
```

## 1  Independent-Samples *t*-Test

**Research question:** Do control and treatment participants differ in reaction time?

```python
ctrl = rt.loc[rt['condition'] == 'control',    'reaction_time_ms']
trt  = rt.loc[rt['condition'] == 'treatment',  'reaction_time_ms']

result = aem.t_test(ctrl, trt)
print(result.to_string())
```

```python
fig = aem.plot_means(rt, x='condition', y='reaction_time_ms',
                     title='Reaction Time by Condition')
plt.show()
```

## 2  One-Way ANOVA

**Research question:** Does encoding strategy affect recall score?

```python
groups = [mem.loc[mem['strategy'] == s, 'recall_score'] for s in mem['strategy'].unique()]
anova  = aem.anova_one_way(*groups)
print(anova.to_string())
```

```python
fig = aem.plot_means(mem, x='strategy', y='recall_score',
                     title='Mean Recall Score by Encoding Strategy')
plt.show()
```

## 3  Interaction Plot (Caffeine × Time of Day)

```python
fig = aem.plot_interaction(caf, x='dose', y='alertness', group='time_of_day',
                            title='Alertness by Caffeine Dose and Time of Day')
plt.show()
```

## 4  Pearson Correlation

```python
corr = aem.correlation(rt['age'], rt['reaction_time_ms'])
print(f'r = {corr["r"]:.3f}, p = {corr["p_value"]:.3f}')
```

## 5  Prospective Power Analysis

How many participants per group do we need to detect a medium effect (d = 0.5)
with 80 % power at α = .05?

```python
pa = aem.power_analysis(effect_size=0.5, alpha=0.05, power=0.80)
print(pa.to_string())
print(f'\n→ You need {int(pa["n_per_group"])} participants per group.')
```

```python
# Sensitivity analysis: vary effect size
import pandas as pd
effect_sizes = [0.2, 0.3, 0.5, 0.8, 1.0]
rows = [aem.power_analysis(d) for d in effect_sizes]
pd.DataFrame(rows)
```
