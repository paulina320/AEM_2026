3. Generate synthetic experiment data for practice.

```python
import aem
print('aem version:', aem.__version__)
print('\nAvailable tools:', [x for x in dir(aem) if not x.startswith('_')])
```

## 1  Sample Datasets

```python
# Load the reaction-time dataset
rt = aem.load_sample('reaction_times')
print(rt.shape)
rt.head()
```

```python
rt.groupby('condition')['reaction_time_ms'].describe()
```

## 2  Generating Synthetic Data

```python
synth = aem.generate_experiment_data(
    n_subjects=25,
    conditions=['placebo', 'low_dose', 'high_dose'],
    effect_size=0.6,
    seed=42,
)
synth.groupby('condition')['score'].describe()
```

## 3  Descriptive Statistics

```python
for condition, group in rt.groupby('condition'):
    stats = aem.descriptive_stats(group['reaction_time_ms'])
    print(f'\n--- {condition} ---')
    print(stats.to_string())
```

## 4  Quick Visualisation

```python
import matplotlib.pyplot as plt

fig = aem.plot_means(rt, x='condition', y='reaction_time_ms', title='Mean Reaction Time by Condition')
plt.show()
```

```python
fig = aem.plot_distribution(rt['reaction_time_ms'], label='Reaction Time (ms)')
plt.show()
```
