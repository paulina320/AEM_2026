# AEM 2026 – Notebook 2: Experimental Design

This notebook walks through the core experimental-design tools provided by the `aem` package:

| Function | Purpose |
|---|---|
| `full_factorial` | All treatment combinations |
| `randomize_order` | Shuffle trial order |
| `latin_square` | Counterbalanced scheduling |
| `block_design` | Randomised complete-block design |
| `counterbalance` | All possible orderings of conditions |

```python
import aem
import pandas as pd
```

## 1  Full-Factorial Design

```python
# 2 × 2 design: Instruction type × Feedback
design = aem.full_factorial({
    'instruction': ['explicit', 'implicit'],
    'feedback':    ['immediate', 'delayed'],
})
print(f'{len(design)} treatment combinations')
design
```

```python
# Randomise trial order for participant 1
p1_design = aem.randomize_order(design, seed=101)
p1_design
```

## 2  Latin Square

```python
import numpy as np

# 4 conditions to be presented in 4 different orders across 4 subjects
ls = aem.latin_square(4, seed=7)
conditions = ['A', 'B', 'C', 'D']

df_ls = pd.DataFrame(
    [[conditions[c] for c in row] for row in ls],
    index=[f'Participant {i+1}' for i in range(4)],
    columns=[f'Period {t+1}' for t in range(4)],
)
print('Latin square (rows=participants, columns=time-periods):')
df_ls
```

## 3  Randomised Complete Block Design

```python
# 3 treatments × 5 blocks
bd = aem.block_design(['control', 'treatment_A', 'treatment_B'], n_blocks=5, seed=0)
print('Block design (15 runs total):')
bd
```

## 4  Counterbalancing

```python
# All possible orderings of 3 conditions
orders = aem.counterbalance(['Task A', 'Task B', 'Task C'])
print(f'{len(orders)} orderings:')
for i, order in enumerate(orders, 1):
    print(f'  Group {i}: {" → ".join(order)}')
```
